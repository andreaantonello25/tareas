### Ejercicio 1

Este ejercicio lo realizamos en clases. Teníamos que organizar números desde el -100 hasta el 200.  

Algo muy interesante es que tuvimos que poner hasta el número **201**, porque en `range`, Python **no incluye el número final**.  

Entonces, si poníamos solamente: "for i in range(-100, 200):", la secuencia llegaría desde el -100 hasta el 199.


```python
#ejercicio 1 tarea 2
for i in range (-100,201):
  print(i)
```

### Ejercicio 2

El segundo ejercicio también lo realizamos en clases.  
Aquí lo que teníamos que hacer, basándonos en el ejercicio 1, era tomar esa misma secuencia que creamos pero ahora hacer que solo imprima los números que son divisibles entre **6** con **resto 0**.

Aprendimos a usar el operador de **resto (`%`)** y cómo `==` se usa para hacer una **comparación igualada**, a diferencia de `=`, que **asigna un valor**.

```python
#ejercicio 2 tarea 2
for z in range (-100,201):
  if z % 6 == 0:
    print(z)
```

### Ejercicio 3

En este teníamos que realizar un código que me sume dos números y me diga si la suma es:

- “menor a 200”  
- “mayor a 200 y menor a 250”  
- “mayor a 250”

Un poco más abajo van a poder ver el “ejercicio 3.1”. Esto es porque el primero cumple exactamente con las instrucciones entregadas, pero decidí hacer el segundo para que esté un poco más ordenado y, más importante, para que también tome en cuenta la posibilidad de que **la suma sea igual a 200** o que **la suma sea igual a 250**.

Esta diferencia la hice porque me pareció pertinente, ya que en el primer código, cuando coloco 200 o 250, me dice que es “mayor” que ese número, cuando en verdad es **igual** al número.

En este código utilicé: `int`, `input`, y las variantes de `if`, `elif` y `else`, junto con `print`.

```python
#ejercicio 3 tarea 2

##pedir los dos números
numero1 = int(input("ingrese el primer numero: "))
numero2 = int(input("ingrese el segundo numero: "))

suma = numero1 + numero2

if suma < 200:
  print("menor a 200")
elif suma >= 200 and suma <= 250:
  print("mayor a 200 y menor a 250")
else:
  print("mayor a 250")
```

```python
#ejercicio 3.1 tarea 2

##Pedir los dos números
numero1 = int(input("Ingrese el primer número: "))
numero2 = int(input("Ingrese el segundo número: "))

suma = numero1 + numero2

if suma < 200:
  print("La suma es menor a 200")
elif suma == 200:
  print("La suma es igual a 200")
elif suma > 200 and suma < 250:
  print("La suma es mayor a 200 y menor a 250")
elif suma == 250:
  print("La suma es igual a 250")
else:
  print("La suma es mayor a 250")
```

### Ejercicio 4

En este teníamos que crear un código que contenga una **función**, en donde me pregunte por la **edad** y el **gusto por la música urbana**, y dependiendo de la respuesta te diga si tienes **más o menos de 18 años**, y si te **gusta o no** la música urbana.

Aquí también realicé un segundo código llamado “ejercicio 4.1”, por las mismas razones del anterior: para que si tienes 18 años te diga que tienes **exactamente esa edad**, y no que tienes menos de 18 años.

Volví a utilizar `int` e `input`, así como `if`, `elif` y `else`, estos dentro del `return`.  
También se agrega el uso de `def`, la cual **sirve para definir una función**.

```python
#ejercicio 4 tarea 2

#Pedir datos al usuario
edad = int(input("Ingrese su edad: "))
gusto = int(input("¿Te gusta la música urbana? (Sí = 1, No = 2): "))

##Función que evalúa edad y gusto por musica ubana
def evaluar_usuario(edad, gusta_musica_urbana):
    if edad <= 18 and gusta_musica_urbana == 1:
        return "Tienes menos de 18 años y te gusta la música urbana"
    elif edad <= 18 and gusta_musica_urbana == 2:
        return "Tienes menos de 18 años y no te gusta la música urbana"
    elif edad > 18 and gusta_musica_urbana == 1:
        return "Tienes más de 18 años y te gusta la música urbana"
    else:
        return "Tienes más de 18 años y no te gusta la música urbana"

resultado = evaluar_usuario(edad, gusto)
print(resultado)
```
